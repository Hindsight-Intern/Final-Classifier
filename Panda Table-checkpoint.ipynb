```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

urls = []
url = 'https://www.cbsnews.com/feature/2021-olympics-summer-tokyo/#'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
#gets every single link as an output
urls = []
for link in soup.find_all('a'):
    articles = link.get('href')
    if (articles.startswith( 'https://www.cbsnews.com/news/' )) == True:
        urls.append(articles)
```

```python
titles = []
for link in soup.find_all('a'):
    articles = link.get('href')
    if (articles.startswith( 'https://www.cbsnews.com/news/' )) == True:
        html_content = requests.get(link.get('href')).text
        soup = BeautifulSoup(html_content, "lxml")
        text = soup.title.text
        titles.append(text)
```

```python
from transformers import pipeline
# testing out the pipeline with the correct labels
main_tag = []
main_score = []

classifier = pipeline("zero-shot-classification")
candidate_labels = ["politics", "international news", "celebrity", "sports", "health", "nutrition", "fitness", "beauty", "business", "economy", "finance", "technology", "science", "lifestyle"]
```

```python
counter = 0
print('we gotta wait until the counter hits ' + str(len(titles)))
# the next step takes forever so i made a counter
```

```python
for i in titles:
    res = classifier(i, candidate_labels, multi_label=True)
    labels = res['labels'] 
    scores = res['scores']
    main_tag.append(labels[0])
    main_score.append(scores[0])
    counter = counter + 1
    print(counter)
```

```python
pd.set_option('display.max_rows', None)
df = pd.DataFrame({'url' : urls,'title' : titles, 'tag': main_tag, 'likelihood': main_score})
display(df)
```
